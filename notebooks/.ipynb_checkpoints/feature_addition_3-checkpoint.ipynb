# Feature Selection and Engineering

## Steps

Step 1: Run a Random Forest to Assess Feature Importance
Why?
Running a random forest classifier provides a quick and interpretable way to rank features by their predictive power on churn. This helps prioritize features before deciding on encoding or combining.

How?

Encode categorical variables temporarily (e.g., one-hot encoding or label encoding).
Fit a random forest classifier on the dataset.
Extract feature importance (feature_importances_ in scikit-learn).
Outcome:
You’ll identify:

Features with high importance that you should retain.
Low-importance features that can be considered for removal or combination.
Step 2: Target Encoding for High-Cardinality Features
Why?
For categorical features with a large number of unique values (e.g., PaymentMethod or InternetService), target encoding can effectively capture their relationship with churn without creating too many dimensions like one-hot encoding.

How?
Replace categorical values with the mean churn rate for each category. For example:

If InternetService = Fiber optic, encode as the mean churn rate of customers with Fiber optic.
Outcome:
Encoded features directly capture the relationship with churn.

Step 3: Combine Related Features
Why?
Features with high Chi-Square dependence or domain-relevant relationships (e.g., InternetService and OnlineSecurity) may add redundancy or noise if kept separate. Combining them simplifies the data and can enhance model performance.

How?
Create interaction features:

Combine InternetService and OnlineSecurity to form categories like Fiber_Without_Security.
Combine Contract and PaperlessBilling to form Billing_and_Contract.
Outcome:
A smaller, more meaningful feature set.

Suggested Workflow
Random Forest:
Determine which features are crucial for churn prediction.
Target Encoding:
Apply to high-cardinality categorical features.
Feature Combination:
Combine features based on high dependence or domain logic.
Re-evaluate:
Check correlation and significance of the new features to ensure redundancy is reduced.

- non numeric columns need to be encoded

```python
numeric_cols = ['tenure','MonthlyCharges','TotalCharges','tenurePerTotalCharges',
                'monthlyChargesPerTotalCharges','totalChargePerMonthlyCharges']
train_str = train.drop(numeric_cols,axis=1)

for col in train_str.columns:
    print(col,train_str[col].unique())
```

```python
#remove columns already numeric
train_str = train_str.drop(['customerID','SeniorCitizen','cluster','cluster','high_risk_customer','discount','churn'], axis=1)
for col in train_str.columns:
    print(col,train_str[col].unique())
```

```python
#Encode string columns
#initialize encoder that drops one level and returns a non sparse matrix
encoder = OneHotEncoder(drop='first', sparse=False)
#encode string columns which returns a non sparse array
str_to_num_cols = encoder.fit_transform(train_str)
#return df with the column names
str_df = pd.DataFrame(str_to_num_cols, columns=encoder.get_feature_names_out())
#add back to df
enc_train = pd.concat([train, str_df], axis=1).drop(train_str.columns, axis=1)

enc_train.head()                   
#
```

```python
#For tree based models, not distance based models there is no need to scale features
#Devide encoded_train into training and validation
target = enc_train['churn']
X_train, X_val, y_train,y_val = train_test_split(enc_train.drop(['customerID','churn'],axis=1),target,test_size=0.3, random_state=42)
X_train.shape, X_val.shape, y_train.shape,y_val.shape
```

```python
rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train,y_train)
```

```python
feature_imp_df = pd.DataFrame(rfc.feature_importances_, index=X_train.columns).reset_index()
feature_imp_df.columns = ['feature','feat_imp']
feature_imp_df = feature_imp_df.sort_values(by='feat_imp').reset_index()
plt.figure(figsize=(10,10))
plt.barh(feature_imp_df['feature'],feature_imp_df['feat_imp'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
plt.show()
```

```python
preds = rfc.predict(X_val)
roc_auc = roc_auc_score(y_val, preds) 
prec_rec = average_precision_score(y_val, preds)  
f1 = f1_score(y_val, preds) 
recall = recall_score(y_val, preds) 
precision = precision_score(y_val, preds) 
print(f'roc_auc-scr = {roc_auc},\nprec_rec_scr= {prec_rec},\nf1_scr= {f1},\nrecall={recall}, \nprecision= {precision}')
```

```python
#add the scores to a df
scores_df = pd.DataFrame(columns=['model','roc_auc','prec_rec','f1','recall','precision'])
scores_df.loc[len(scores_df)] = ['org_data',roc_auc,prec_rec,f1,recall, precision]
scores_df
```

The model performs poorly, the precision_recall score is below 50%. To test importance of features:

ways to try to reduce features
1. features with importance < 0.01 will be removed then test the model
2. Try recursive feature elimination to see the number of features that yield the best scores
2. features with importance between 0.01 and 0.02 will be target encoded to help capture their relationship with the target
- to do this two functions will be created to create the train test sets and one to run the model and return scores. After each change feature importance will be returned.

### Remove least important features:
- Use Recursive Feature Elimination (RFE) which recursively removes the least important features and evaluates model performance. It works also well with features that are correlated.
- RFE considers feature interactions and model performance, making it more reliable than raw importance scores

```python
#1. features with importance < 0.01 will be removed then test the model

non_imp_feat = list(feature_imp_df[feature_imp_df['feat_imp'] <0.01]['feature'])
#drop these features from enc_train
new_train = X_train.drop(non_imp_feat, axis=1)
new_val = X_val.drop(non_imp_feat, axis=1)
len(X_train.columns) - len(new_train.columns) 
```

```python
new_train.columns
```

```python
#2. Try recursive feature elimination and measure the scores and the number of features that yield the best scores 

model = RandomForestClassifier(random_state=42)

num_features = []
auc_roc_scores = []
avg_precision = []
f1_list = []
recalls = []

for n_features in range(5, X_train.shape[1],5):
    rfe = RFE(estimator=model, n_features_to_select=n_features)
    rfe.fit(X_train, y_train)
    X_train_rfe = rfe.transform(X_train)
    X_test_rfe = rfe.transform(X_val)
    
    
    model.fit(X_train_rfe,y_train)
    y_pred_proba = model.predict_proba(X_test_rfe)[:,1]
    y_preds = model.predict(X_test_rfe)
    
    auc = roc_auc_score(y_val,y_pred_proba)
    prec_avg = average_precision_score(y_val,y_pred_proba)
    f1 = f1_score(y_val,y_preds)
    recall_scr = recall_score(y_val,y_preds)
    
    #store results
    num_features.append(n_features)
    auc_roc_scores.append(auc)
    avg_precision.append(prec_avg)
    f1_list.append(f1)
    recalls.append(recall_scr)
```

- ROC_AUC plots true positives to false positives, it reflects the model's ability to distinguish between classes
- average precision summarizes the precision_recall curve. It is suitable for imbalanced data which this data is
- f1 score balances precision and recall

```python
#plot roc_auc curve, average precision and f1 score
plt.figure(figsize=(10,10))
plt.plot(num_features,auc_roc_scores, label='ROC_AUC')
plt.plot(num_features,avg_precision,label='AVG_PRECISION')
plt.plot(num_features,f1_list, label='F1 Score')
plt.plot(num_features,recalls, label='Recall')
plt.xlabel('Number of features')
plt.ylabel('Score')
plt.title('Evaluation metrics vs. number of features using RFE')
plt.grid()
plt.legend()
print(f'max num of features = {num_features[np.argmax(auc_roc_scores)]} with auc_roc = {max(auc_roc_scores)} \n\
max num of features = {num_features[np.argmax(avg_precision)]} with average precision = {max(avg_precision)}\n\
max num of features = {num_features[np.argmax(f1_list)]} with f1 score = {max(f1_list)}\n\
max num of features = {num_features[np.argmax(recalls)]} with recall = {max(recalls)} ')
```

```python
#plot roc_auc curve, average precision and f1 score
plt.figure(figsize=(10,10))
plt.plot(num_features,auc_roc_scores, label='ROC_AUC')
plt.plot(num_features,avg_precision,label='AVG_PRECISION')
plt.plot(num_features,f1_list, label='F1 Score')
plt.plot(num_features,recalls, label='Recall')
plt.xlabel('Number of features')
plt.ylabel('Score')
plt.title('Evaluation metrics vs. number of features using RFE')
plt.grid()
plt.legend()
print(f'max num of features = {num_features[np.argmax(auc_roc_scores)]} with auc_roc = {max(auc_roc_scores)} \n\
max num of features = {num_features[np.argmax(avg_precision)]} with average precision = {max(avg_precision)}\n\
max num of features = {num_features[np.argmax(f1_list)]} with f1 score = {max(f1_list)}\n\
max num of features = {num_features[np.argmax(recalls)]} with recall = {max(recalls)} ')
```

Since I am looking for early signals for customer churn I need to minimize missed positive cases or churned customers ==> meaning false negatives.

- Recall is good if we want to capture the max number of positives even if at the cost of  a few false positives, 
- average precision is great for testing different thresholds for the model's performance for imbalanced data
- f1 score gives a balanced approach if intervention for false positives are expensive
- Since I don't want to miss churned cases, recall is a good metric to use, and average precision score shows how well positive cases are ranked across thresholds. To be safe I will start with 25 features and then test the thresholds  

```python
#test thresholds
#data with only 25 features
model = RandomForestClassifier(random_state=42)

rfe = RFE(estimator=model, n_features_to_select=25)
rfe.fit(X_train, y_train)
X_train_25 = rfe.transform(X_train)
X_val_25 = rfe.transform(X_val)
len(X_train_25[0]),len(X_train_25)

```

```python
#features used are 
X_train.columns[rfe.support_]
```

```python
#fit the model on the new data
model.fit(X_train_25,y_train)
#predict probabilities the positive class
pred_prob = model.predict_proba(X_val_25)[:,1]
#predict
y_preds = model.predict(X_val_25)
#compute precisition recall curve
precision, recall, thresholds = precision_recall_curve(y_val, pred_prob)

f1_scores = 2*(recall*precision) / (precision+recall+1e-10) #1e-10 so no division by 0

plt.figure(figsize=(10,10))
plt.plot(thresholds, precision[:-1],c='blue',label='precision',linestyle='--')
plt.plot(thresholds,recall[:-1],label='recall',linestyle='--',c='red')
plt.plot(thresholds,f1_scores[:-1],label='f1 score',linestyle='--',c='black')
plt.axhline(f1_scores[f1_scores.argmax()],label='max f1 score')
plt.legend()
plt.grid()
plt.title('Precision, recall, and f1 score vs. threshold')
plt.xlabel('Thresholds')
plt.ylabel('Score')
plt.show()
```

- I want to capture potential churners which means higher recall but at the cost of more false positives (non churners falsely labeled as churners). 
- The best threshold is where at the max point of f1 score, at this point recall > precision.
- This will serve as an early warning system at the cost of some false alarms. It will also prevent loss of revenue by losing customers. If the cost of false positives becomes too high then the threshold can be adjusted


```python
#max_f1score = f1_scores[f1_scores.argmax()]
optimal_idx = np.argmax(f1_scores)
optimal_threshold = thresholds[optimal_idx]

#reevaluate the predictions
y_preds_optimized = (pred_prob >= optimal_threshold).astype(int)
#print metrics for without and with optimization
precision_org = precision_score(y_val,y_preds)
precision_opt = precision_score(y_val,y_preds_optimized)

recall_org = recall_score(y_val,y_preds)
recall_opt = recall_score(y_val,y_preds_optimized)

f1_org = f1_score(y_val,y_preds)
f1_opt = f1_score(y_val,y_preds_optimized)

roc_org = roc_auc_score(y_val,y_preds)
roc_opt = roc_auc_score(y_val,y_preds_optimized)

print(f'original score precision={precision_org} optimized precision={precision_opt}\n\
original recall = {recall_org} optimized = {recall_opt}\n\
orginal f1 score = {f1_org} optimized = {f1_opt}\n\
original roc score = {roc_org} optimized = {roc_opt}')
```

```python
scores_df = pd.DataFrame(columns = ['model',' ' ,'precision','recall','f1','roc'])
scores_df.loc[len(scores_df)] = ['randomforest','original_score',precision_org,recall_org,f1_org,roc_org]
scores_df.loc[len(scores_df)] = ['randomforest','after_threshold_score',precision_opt,recall_opt,f1_opt,roc_opt]
scores_df
```
