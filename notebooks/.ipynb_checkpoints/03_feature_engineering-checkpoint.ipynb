## Chi squared test for features
- Now lets create something like corr for categorical vars
    - create a list of all possible combinations of all categorical variables
    - run chi squared on all categrical columns and include churn

```python
def chi_squared_test(data,cols):
    '''
    - creates a cross table of all combinations of categorical features
    - calculates chi squared score for each combination
    *******************************
    parameters:
        full data set, 
        categorical columns
    *******************************
    return:
        list of all combinations with their chi squared 
    '''
    
    cat_cols_1 = cols
    cat_cols_2 = cols

    cat_prod = list(product(cat_cols_1,cat_cols_2)) #create a product of columns

    result = []
    for i in cat_prod:
        
        if i[0]!=i[1]:
            cross_tab = pd.crosstab(data[i[0]],data[i[1]])
            result.append((i[0],i[1],chi2_contingency(cross_tab)[1]))
   
    return result
```

```python
#categorical columns
cat_cols_1 = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
           'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
           'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod',
           'cluster_4','discount','Churn']
chi_results = chi_squared_test(train_data, cat_cols_1)
```

```python
chi_df = pd.DataFrame(chi_results,columns = ['col1','col2','pval'])
#pivot df 
chi_df = chi_df.pivot(index='col1',columns='col2',values='pval').sort_values(by='Churn')
chi_df
```

### Commitment score feature can be created using a combination of these 3 features:
- If a user has a two_year contract, payes using an automatic bank transfer or credict card that is `high` commitement score
- If a user has one_year and payes the same way that is `med_high`
- If a user has month_to_month payment but automatic payment that is `med_low`
- If a user has month_to_month and  Mailed or electronic check then that is `low`


- Create a df with contract, paymentMethod and paperlessBilling

```python
payment = train_data[['Contract','PaymentMethod','PaperlessBilling']]
payment.head()
```

```python
def commitment_score(row):
    '''
    Asses commitment based on contract and payment method
    
    parameters: a row in the dataframe
    return: commitment score [high, med_high, med_low, low]
    '''
    
    if row['Contract'] == 'Two year':
        if row['PaymentMethod'] in ['Bank transfer (automatic)','Credit card (automatic)']:
            return 'high'
        else:
            return 'med_high'
    elif row['Contract'] == 'One year':
        if row['PaymentMethod'] in ['Bank transfer (automatic)','Credit card (automatic)']:
            return 'med_high'
        else:
            return 'med_low'
    else:
        if row['PaymentMethod'] in ['Bank transfer (automatic)','Credit card (automatic)']:
            return 'med_low'
        else:
            return 'low'


payment['commitment_level'] = payment.apply(commitment_score,axis=1)
payment.head()
```

```python
#add it to df
train_data['commitment_level'] = payment['commitment_level']
```

### Risk score:
- contract: month_to_month has the highest risk score of 3, one_year is 2 and two years = 1
- paymentMethod: Electronic check has the highest risk at 4, Mailed check is 3, Credit card (automatic) is 2 and Bank transfer (automatic) at 1 because automatic Bank transfer is the the most committed
- PaperlessBilling: yes is 2 no is one

```python
def risk_score_calculator(row):
    '''
    calculates the risk score according to customers' payment method
    '''
    risk_score = 0
    #Contract
    if row['Contract'] == 'Month-to-month':
        risk_score = 3
    elif row['Contract'] == 'One year':
        risk_score = 2
    #if two_year then 1
    else:
        risk_score = 1
    #PaymentMethod
    if row['PaymentMethod'] == 'Electronic check':
        risk_score += 4
    elif row['PaymentMethod'] == 'Mailed check':
        risk_score += 3
    elif row['PaymentMethod'] == 'Credit card (automatic)':
        risk_score += 2
    #if Bank transfer (automatic) then 1
    else:
        risk_score += 1
    #PaperlessBilling
    if row['PaperlessBilling'] == 'Yes':
        risk_score +=1
    # if PaperlessBilling = no then 0
    return risk_score
    
payment['payment_risk_score'] = payment.apply(risk_score_calculator, axis=1)
payment.head()
```

```python
def family_structure(row):
    '''returns family structure for partner and dependents columns
    '''
    if row['Partner'] == 'Yes':
        if row['Dependents'] == 'Yes':
            status = 'family'
        else: # has a partner but no dependents
            status = 'couple'
    else: # has no partner
        if row['Dependents'] == 'Yes':
            status = 'single_parent'
        else: # not partner and no dependents
            status = 'single'
    return status

```

### Combine continuous features (tenure, monthlyCharges, totalCharges)
- historicalAvgSpending: totalCharges/tenure = total amount customers have paid / how long they have been a customer ==> historical average spending
- spendingTrend: compares monthlyCharges which is the current spending to their historicalAvgSpending, if tenure = 0 then the value will be monthlyCharges else totalCharges/tenure:
  
      * if their spending ration = 1.0 ==> the customer is spending the same what they have spent on averge over their tenure
        
      * greater than 1.0 means they are spending more than their historical average: example 1.25 then they are spending 25% more    
           - customers with speninding trend > 1.0 means than they are satisfied and have lower churn risk
  
      * less than 1.0 then the customer is spending less than their historical average, example 0.8 means they are spending 20% less
          - customer with spending trend < 1.0 means that they are cancelling service or are negotiating lowering their charges and this indicates dissatisfaction and the increased possibility of churn

```python
def extract_target(data):
    '''
    input: cleaned data
    output: maps target values with 1 and 0 and returns the mapped target
    '''
    y = data['Churn'].map({'Yes':1,'No':0})
    return y
    
def scale_encode_data(data):
    '''
    Transforms data by scaling numeric features and encoding non_numeric features
    
    Parameters: cleaned_data
    Returns: scaled data with customerID and Churn dropped
    '''

    #map binary features 
    data['gender'] = data['gender'].map({'Male':1,'Female':0})
    data['Partner'] = data['Partner'].map({'Yes':1,'No':0})
    data['Dependents'] = data['Dependents'].map({'Yes':1,'No':0})
    data['PhoneService'] = data['PhoneService'].map({'Yes':1,'No':0})
    data['PaperlessBilling'] = data['PaperlessBilling'].map({'Yes':1,'No':0})
    
    # non numeric columns
    types = ['object','category']
    #categ_cols = ['gender','Partner','Dependents','PhoneService','PaperlessBilling']
    non_num_cols = []
    for col in data.columns:
        if data[col].dtype in types:
            non_num_cols.append(col)
    print(non_num_cols)
    
    non_num_data= data[non_num_cols]
    non_numeric_dummies =pd.get_dummies(data[non_num_cols], drop_first=True)#get_dummies gives  binary features
    #change values to numeric 
    dummies_num = non_numeric_dummies.astype(int)
    

    #concat to original dataframe and drop non-numeric original columns
    df_num = pd.concat([data.drop(non_num_cols,axis=1),dummies_num],axis=1)
    df_num = df_num.reset_index(drop = True)
    #df_num_all = pd.concat([df_num, data[categ_cols]],axis=1)
    
    #scale numeric features
    cols_to_drop = non_num_cols
    cols_to_drop.extend(['gender','Partner','Dependents','PhoneService','PaperlessBilling'])
    
    df_to_scale = data.drop(cols_to_drop, axis=1)
    
    #num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

        
    scaler=StandardScaler()
    #scale 
    scaled_num = pd.DataFrame(scaler.fit_transform(df_to_scale),columns=df_to_scale.columns)
    
    #drop old non scaled features and concatinate the new scaled features
    df_scaled = pd.concat([df_num.drop(df_to_scale.columns, axis=1),scaled_num],axis=1)
   
    #drop Churn and customerID
    final_df = df_scaled.drop(['customerID','Churn'], axis=1)
    
    return final_df
    
```

```python
class CustomFeatureProcessor(BaseEstimator, TransformerMixin):
    '''
    Create a custom transformer for the scale_encode_data function by
    inheriting from the two base classes to create a scikit learn compatible transformer, 
    they provide necessary methods to work with the pipeline API
    
    '''
    def __init__(self, processing_func):
        self.processing_func = processing_func

    def fit(self, X, y=None):
        #y=None allows the transformer to be used in supervised and unsupervised 
        return self

    def transform(self, X, y=None):
        # should handle data processing and return X, y
        X_processed = self.processing_func(X)
        return X_processed
```

```python
#Process data
def evaluate_model(data, processing_func):
    '''
    Evaluate RandomForest using a custom processing function

    parameters: 
        pandas df: cleaned IBM churn data
        feature processing function that takes the data and returns scaled, encoded data
    Returns:
        dict: cross-validation results
    '''
    #target variable
    y = extract_target(data)
    
    #create a pipeline with feature processing
    pipeline = Pipeline([
                    ('processor',CustomFeatureProcessor(processing_func)),
                    ('classifier',RandomForestClassifier(random_state=42))
        ])

    #stratified k-fold
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    scoring = {'precision':make_scorer(precision_score),
                'recall':make_scorer(recall_score),
                'f1_score':make_scorer(f1_score),
              'auc_score':make_scorer(roc_auc_score)
              } 
    
    #cross validation
    cv_results = cross_validate(pipeline, data, y, cv=skf, scoring=scoring)

    
    return cv_results
```

```python
def metric_calcualation(cv_scores,metrics_dict, colname):
    '''
    Adds a new row of metrics to the metrics dataframe
    parameters:
    the scores from running the model on the dataset, the dataframe where all the metrics are stored, and the number of the features added
    returns:
    the metrics dataframe with the new row added
    '''
    new_row = {
        'feature_set':'org_data',
        'precision_mean':float(round(np.mean(cv_scores['test_precision']),4)),
        'precision_std':float(round(np.std(cv_scores['test_precision']),4)),
        'recall_mean':float(round(np.mean(cv_scores['test_recall']),4)),
        'recall_std':float(round(np.std(cv_scores['test_recall']),4)),
        'f1score_mean':float(round(np.mean(cv_scores['test_f1_score']),4)),
        'f1score_std' : float(round(np.std(cv_scores['test_f1_score']),4)),
        'AUC_mean' : float(round(np.mean(cv_scores['test_auc_score']),4)),
        'AUC_std' : float(round(np.std(cv_scores['test_auc_score']),4)) 
         }
    idx = len(features_metrics)+1
    for k,v in new_row.items():
        features_metrics.loc[idx,k]=v
    return features_metrics
```

- function to plot precision recall and f1 vs. threshold plot and calculate the scores and add them to score_dd

```python
def prec_recall_plot(fitted_model,X_test, y_test,plot=False):


    #predict probabilities the positive class
    pred_prob = fitted_model.predict_proba(X_test)[:,1]
    #predict
    y_preds = fitted_model.predict(X_test)
    #compute precisition recall curve
    precision, recall, thresholds = precision_recall_curve(y_test, pred_prob)
    
    f1_scores = 2*(recall*precision) / (precision+recall+1e-10) #1e-10 so no division by 0
    
    f1_org = f1_score(y_test,y_preds)
    roc_org = roc_auc_score(y_test,y_preds)
    precision_scr = precision_score(y_test,y_preds)
    recall_scr = recall_score(y_test,y_preds)
    
    #plot
    if plot:
        plt.figure(figsize=(10,10))
        plt.plot(thresholds, precision[:-1],c='blue',label='precision',linestyle='--')
        plt.plot(thresholds,recall[:-1],label='recall',linestyle='--',c='red')
        plt.plot(thresholds,f1_scores[:-1],label='f1 score',linestyle='--',c='black')
        plt.axhline(f1_scores[f1_scores.argmax()],label='max f1 score')
        plt.legend()
        plt.grid()
        plt.title(f'Precision, recall, and f1 score vs. threshold')
        plt.xlabel('Thresholds')
        plt.ylabel('Score')
        plt.show()

    
    return precision_scr, recall_scr, roc_org, f1_scores, thresholds, f1_org

```

```python
def scores_calculation(fitted_model, X_test,y_test): 
    
    org_precision,org_recall, org_roc, org_f1_scores, thresholds, org_f1 = prec_recall_plot(fitted_model, X_test,y_test)
   
    #optimal threshold
    optimal_idx = np.argmax(org_f1_scores)
    optimal_threshold = thresholds[optimal_idx]
    
    #predict probabilities the positive class
    pred_prob = fitted_model.predict_proba(X_test)[:,1]
    #predict
    y_preds = fitted_model.predict(X_test)
    #reevaluate the predictions
    y_preds_optimized = (pred_prob >= optimal_threshold).astype(int)
    #print metrics for without and with optimization
    
    precision_opt = precision_score(y_test,y_preds_optimized)


    recall_opt = recall_score(y_test,y_preds_optimized)
    
    f1_opt = f1_score(y_test,y_preds_optimized)

    roc_opt = roc_auc_score(y_test,y_preds_optimized)

    print(f'\
    optimal threshold = {optimal_threshold}\n\
    original precision={org_precision}, optimized precision={precision_opt}\n\
    original recall = {org_recall}, optimized = {recall_opt}\n\
    orginal f1 score = {org_f1}, optimized = {f1_opt}\n\
    original roc score = {org_roc}, optimized = {roc_opt}')
    
    #print(classification_report(y_test, y_preds))
    #print(classification_report(y_test,y_preds_optimized))
    
    return precision_opt, recall_opt, roc_opt, f1_opt, optimal_threshold
```

```python
def add_to_scores(scores_df,fitted_model,name, X_test,y_test):
    
    org_precision,org_recall,roc_org,_,_, org_f1_scores = prec_recall_plot(fitted_model, X_test,y_test)
    precision_opt, recall_opt, roc_opt, f1_opt,_ = scores_calculation(fitted_model, X_test,y_test)
    
    scores_df.loc[len(scores_df)] = [name,'original_score',org_precision,org_recall,org_f1_scores,roc_org]
    scores_df.loc[len(scores_df)] = [name,'after_threshold_score',precision_opt,recall_opt,f1_opt,roc_opt]
    return scores_df
```

### hyperparameter tuning for xgboost


```python
from scipy.stats import randint
param_grid = {'n_estimators':np.arange(100,500,10),#number of trees(boosting rounds), should use early stopping
             'learning_rate':np.arange(0.01,0.3,0.01),#the step size for boosting round
             'max_depth':range(3,10),
             'min_child_weight':range(1,10),#min sum of instances weight needed to make a split
             'gamma':np.arange(0,1),#min loss reduction to make a split, for regularization
             'subsample':np.arange(0.5,1,0.1),#fraction of sample used for each boosting round, prevents overfitting
             'reg_alpha':np.arange(0,1,0.1),#l2 regularization
             'reg_lambda':np.arange(0,1,0.1),#l1 regularization
             'colsample_bytree':np.arange(0.,1,0.01)#fraction of features used for each boosting round
             }
xgboost = XGBClassifier(random_state=42)
rs = RandomizedSearchCV(xgboost, 
                        param_distributions = param_grid, 
                        n_iter=100,
                        cv=5,
                        scoring='f1',
                        n_jobs=-1,
                        random_state=42)
rs.fit(X_train,y_train)
```
