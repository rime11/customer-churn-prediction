## Purpose of project
This project investigates how strategic feature engineering can significantly enhance data precision and model accuracy so that customer churn signals are identified early in the customer life cycle and that interventions can be made before customers decide to leave.

## Question that is being answered:  
"How early can we predict customer churn, and which features are most indicative of early churn risk?"

## The data:
This sample dataset that tracks a fictional telco company's customer churn based on various factors.The churn column indicates whether the customer departed within the last month. Other columns include gender, dependents, monthly charges, and many with information about the types of services each customer has.

- this is to have a baseline value to compare to to measure how much feature addition improves the model
  #### Steps:
- create a custome feature feature processor
- wrap it in a CustomFeatureProcessor  
  - This class that will have a fit and transform methods so that it is compatible with scikit learn's pipeline method. This class is important when I am running cross validation where scikit learn calls fit and transform on each fold. It also prevents data leakage between the test set and the validation sets and makes the function compatible with hyper parameter tuning tools like GridSearch and allows for tracking parameters
- use Stratified k-fold which insures that the class distribution in the test and validation sets matches that of the original data, which kfold validation does not. Since the dataset is imbalaced this will creates more representative samples
- Cross validation


Load Data

```python
train_df_org = pd.read_pickle('org_train.pkl')
train_df_org.head(2)
```

- create a function that encodes and scales the data that I can reuse

```python
def extract_target(data):
    '''
    input: cleaned data
    output: maps target values with 1 and 0 and returns the mapped target
    '''
    y = data['Churn'].map({'Yes':1,'No':0})
    return y
    
def scale_encode_data(data, add_num_cols=False, add_non_num_col=False, add_num = [],add_non_num=[]):
    '''
    Transforms data by scaling numeric features and encoding non_numeric features
    
    Parameters: cleaned_data
    Returns: scaled data with customerID and Churn dropped
    '''
    non_num_cols = ['MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport','StreamingTV', 'StreamingMovies','Contract', 'PaymentMethod']
    if add_non_num_col:
        non_num_cols.append(add_non_num_col)
    non_numeric_dummies =pd.get_dummies(data[non_num_cols], drop_first=True)#get_dummies gives  binary features
    #change values to numeric 
    dummies_num = non_numeric_dummies.astype(int)

    #concat to original dataframe and drop non-numeric original columns
    df_num = pd.concat([data.drop(non_num_cols,axis=1),dummies_num],axis=1)
    df_num= df_num.reset_index(drop = True)

    #map other binary features 
    df_num['gender'] = df_num['gender'].map({'Male':1,'Female':0})
    df_num['Partner'] = df_num['Partner'].map({'Yes':1,'No':0})
    df_num['Dependents'] = df_num['Dependents'].map({'Yes':1,'No':0})
    df_num['PhoneService'] = df_num['PhoneService'].map({'Yes':1,'No':0})
    df_num['PaperlessBilling'] = df_num['PaperlessBilling'].map({'Yes':1,'No':0})

    #map target feature
    #df_num['Churn'] = df_num['Churn'].map({'Yes':1,'No':0})
    
    #scale numeric features
    num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
    if add_num_cols:
        num_cols.append(add_num)
        
    scaler=StandardScaler()
    #scale 
    scaled_num = pd.DataFrame(scaler.fit_transform(data[num_cols]),columns=data[num_cols].columns)
    #drop old features and concatinate the new scaled features
    df_scaled = pd.concat([df_num.drop(num_cols, axis=1),scaled_num],axis=1)
   
    #drop Churn and customerID
    final_df = df_scaled.drop(['customerID','Churn'], axis=1)
    
    return final_df
```

```python
df = scale_encode_data(train_df_org)
```

```python
df.head()
```

```python
def CustomFeatureProcessor(BaseEstimator, TransformerMixin):
    '''
    Create a custom transformer for the scale_encode_data function by
    inheriting from the two base classes to create a scikit learn compatible transformer, 
    they provide necessary methods to work with the pipeline API
    
    '''
    def __init__(self, processing_func):
        self.processing_func = processing_func

    def fit(self, X, y=None):
        #y=None allows the transformer to be used in supervised and unsupervised 
        return self

    def transform(self, X, y=None):
        X_processed,_ = self.processing_func(X)
        return X_processed
```

```python
#Process data
def evaluate_model(data, processing_func):
    '''
    Evaluate RandomForest using a custom processing function

    parameters: 
        pandas df: cleaned IBM churn data
        feature processing function that takes the data and returns scaled, encoded data
    Returns:
        dict: cross-validation results
    '''
    #target variable
    y = extract_target(train_df_org)
    
    #create a pipeline with feature processing
    pipeline = Pipeline([
                    ('processor',CustomFeatureProcessor),
                    ('classifier',RandomForestClassifier())
        ])

    #stratified k-fold
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    scoring = {'precision':make_scorer(precision_score),
               'recall':make_scorer(recall_score),
               'f1_score':make_scorer(f1_score)} 
    
    #cross validation
    cv_results = cross_validate(pipeline, df, y, cv=skf, scoring=scoring)

    return cv_results
```

## Run KMeans

To investigate the best number of clusters I am going to use the elbow method, silhouette score and Davies-Bouldin index

- elbow method: measures within cluster sum of squares, or inertia. Lower inertia means that the points are closer to their centroid so more compact clusters
- silhouette score:measures how similar an point is to its own cluster. It can be 1 which means the clusters are well separated, 0 means that the distance between clusters is not significant, or -1 means the clusters are badly assigned
- Davies-Bouldin index: measures how well the clusters are separated and how compact they are, the lower it is indicates better clusters -> more compact and more separated 

### Elbow method

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
 
#elbow method
inertias = []
silhouette = []
dbi_scores = []
for i in range(2,11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(df_final)
    #calculate inertia
    inertias.append(kmeans.inertia_)
    #calculate silhouette score
    sil_score= silhouette_score(df_final, kmeans.labels_) 
    silhouette.append(sil_score)
    #calculate davies_bouldin score
    dbi_score = davies_bouldin_score(df_final, kmeans.labels_)
    dbi_scores.append(dbi_score)
```

```python
#plot inertials

fig,axes = plt.subplots(3,1,figsize=(10,8))
plt.title('optimal number of clusters')
axes[0].plot(range(2,11),inertias,linestyle='-', marker='o', color='b')
axes[0].set_title('Elbow method')
axes[0].grid(True)
#axes[0].set_xlabel('Number of clusters')
axes[0].set_ylabel('Inertia (Within cluster sum of squares)')
axes[0].set_xticks(range(2,10))

axes[1].plot(range(2,11),silhouette,linestyle='-', marker='o', color='r')
axes[1].set_title('Silhouette method')
axes[1].grid(True)
axes[1].set_xticks(range(2,10))
#axes[1].set_xlabel('Number of clusters')

axes[2].plot(range(2,11),dbi_scores,linestyle='-', marker='o', color='c')
axes[2].set_title('davies_bouldin method')
axes[2].grid(True)
axes[2].set_xticks(range(2,10))
axes[2].set_xlabel('Number of clusters')
plt.tight_layout()
plt.show()
```

### Influence of tenure
* To check if tenure really influences churn I will perform a t-test and a Mann-Whitney U test.
Mann-Whitney U test is a non-parametric test to determine if two samples come from the same distribution, unlike t-test it does not assume that distributions have a normal distribution which makes it suitable for data that is skewed or has outliers because it compares the rank of values instead of the values themselves, which is the case here. The two distributions are going to be for churned and unchurned customers for cluster 2. If p < 0.05 then there is a significant difference

```python
#perform ttest
from scipy.stats import ttest_ind, mannwhitneyu
#create the two distributions
cluster_2_churn = cluster_2_df[cluster_2_df['churn']==1]['tenure']
cluster_2_not_churn = cluster_2_df[cluster_2_df['churn']==0]['tenure']

#ttest
t_stat, p_val = ttest_ind(cluster_2_churn,cluster_2_not_churn,equal_var=False)#does not assume equal population variance
print(f't_stat = {t_stat}\np_value= {p_val}')
```

```python
#Mann_Whitney U test
u_stat, p_val_u = mannwhitneyu(cluster_2_churn,cluster_2_not_churn, alternative='two-sided') #alternative is alternative hypothesis and two sided means that the distributions of the two populations are not the same 
print(f'u_stat = {u_stat}\np_value = {p_val_u}')
```

Both the ttest and the Mann_Whitney U test have a p_value much smaller than 0.05 which is significant, confirming that tenure significantly impacts churn in cluster 2.

- Now I will do run the t_test and mann_whitney U test for cluster 1 and 2

```python
#create a function
def stats_tests(data,col):
    '''
    group distributions by cluster,
    create churn and not churn series
    print the t_stat and p_value for ttest and mann_whitney U 
    '''
    #create the distributions
    dist_cluster_1 = data[data['cluster']==0]
    
    dist_cluster_2 = data[data['cluster']==1]
    dist_cluster_3 = data[data['cluster']==2]
    #print(f'distribution:\n{dist_cluster_1}')
    #create the churn clusters for distr
    clusters_distr = [dist_cluster_1,dist_cluster_2,dist_cluster_3]
    
    for i,distr in enumerate(clusters_distr, start=1):
        cluster_churn_i = distr[distr['churn']==1][col]
        cluster_not_churn_i = distr[distr['churn']==0][col]
    
        #ttest
        t_stat, p_val_t = ttest_ind(cluster_churn_i,cluster_not_churn_i,equal_var=False)
        #mannwhitneyu test
        u_stat, p_val_u = mannwhitneyu(cluster_churn_i,cluster_not_churn_i, alternative='two-sided') #alternative is alternative hypothesis and two sided means that the distributions of the two populations are not the same 
        
        print(f'For {col}')
        print(f'for cluster_{i}:\nt_stat = {t_stat}\nttest_p_value= {p_val_t}')
        print(f'\nu_stat = {u_stat}\nu_p_value = {p_val_u}\n')
```

```python
#ttest and mannwhitney for tenure
stats_tests(num_df[['tenure','cluster','churn']],'tenure')
```

All p_values for the three clusters for tenure are significant meaning that in all three clusters churned customers are significantly different than non-churned customers
